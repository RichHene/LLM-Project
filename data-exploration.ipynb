```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path) #prints the path to the dataset
```

```python
import pandas as pd #read into pandas
```

```python
books = pd.read_csv(f'{path}/books.csv')
```

```python
books.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar = False, ax = ax)

plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

```python
#know any biasese present/skewed towards any types of books
```

```python
#create a numeric representation of books that are missing
import numpy as np
books['missing_description'] = np.where(books['description'].isna(), 1, 0) #put a 1 in columns where the description is missing otherwise 0
books['age_of_book'] = 2025 - books['published_year'] #convert the year of the book to the age of the book
```

```python
#columns of interest for the book recommender
columns_of_interest = ['num_pages', 'age_of_book', 'missing_description', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method = 'spearman') #generates a correlation matrix


sns.set_theme(style = 'white')
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt = '.2f', cmap = 'coolwarm', cbar_kws={'label': 'Spearman Correlation'})
heatmap.set_title('Correlation Heatmap')
plt.show()
```

```python
#if it is missing fetch me all of those observations, then ~ to negate and do the opposite and keep them
books_missing = books[(~books['description'].isna()) &
        (~books['num_pages'].isna()) &
        (~books['average_rating'].isna()) &
        (~books['published_year'].isna())
]
```

```python
#books_missing['words_in_description'] = books_missing['description'].str.split().str.len()
books_missing.loc[:, 'words_in_description'] = books_missing['description'].str.split().str.len()
```

```python
book_missing_25_words = books_missing[books_missing['words_in_description'] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[ : , 'title_and_subtitle'] = (
    np.where(book_missing_25_words['subtitle'].isna(), book_missing_25_words['title'],
             book_missing_25_words[['title', 'subtitle']].astype(str).agg(": ".join, axis = 1)) #aggregate the subtitle and the title
)
```

```python
book_missing_25_words.head()
```

```python
book_missing_25_words.loc[ : , 'tagged_description'] = book_missing_25_words[['isbn13', 'description']].astype(str).agg(" ".join, axis = 1)
```

```python
#drop the fields that we do not need/modified from
(
    book_missing_25_words
    .drop(['subtitle', 'missing_description', 'age_of_book', 'words_in_description'], axis = 1)
    .to_csv('books_cleaned.csv', index = False)
)
```
